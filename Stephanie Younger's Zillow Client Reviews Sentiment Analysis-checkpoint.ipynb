# Import Necessary Libraries

```python
import pandas as pd
import numpy as np
from nltk.tokenize import word_tokenize
import string
from textblob import TextBlob
import re
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.decomposition import LatentDirichletAllocation
import matplotlib.pyplot as plt
import random
```

```python
Stephanie = pd.DataFrame(Stephanie)
```

 # **Comment Preprocessing** 

```python
# Tokenization, Lowercasing, and Punctuation Removal
def preprocess_comment(comment):
    # Tokenization
    tokens = word_tokenize(comment)
    
    # Lowercasing
    lowercased_tokens = [token.lower() for token in tokens]
    
    # Punctuation Removal
    no_punctuation_tokens = [token for token in lowercased_tokens if token not in string.punctuation]
    
    # Reconstruct the preprocessed comment
    preprocessed_comment = " ".join(no_punctuation_tokens)
    
    return preprocessed_comment

# Apply the preprocessing function to the 'comment' column
Stephanie['comment_preprocessed'] = Stephanie['comment'].apply(preprocess_comment)

# Display the preprocessed DataFrame
print(Stephanie[['comment', 'comment_preprocessed']])
```

```python
display(Stephanie)
```

```python
# Define a function to clean preprocessed comment
def clean_preprocessed_comment(comment):
    # Remove special characters and symbols
    comment = re.sub(r'[^\w\s]', '', comment)
    
    # Correct typos and misspelled words
    blob = TextBlob(comment)
    comment = str(blob.correct())
    
    # Remove unnecessary white spaces
    comment = " ".join(comment.split())
    
    return comment

# Apply the cleaning function to the 'Comment_Preprocessed' column
Stephanie['cleaned_preprocessed_comment'] = Stephanie['comment_preprocessed'].apply(clean_preprocessed_comment)

# Display the cleaned DataFrame
print(Stephanie[['comment_preprocessed', 'cleaned_preprocessed_comment']])
```

```python
display(Stephanie)
```

# Aspect Identification 
Reference - [Anoop, V.S. and Asharaf, S.. "Aspect-Oriented Sentiment Analysis: A Topic Modeling-Powered Approach" Journal of Intelligent Systems, vol. 29, no. 1, 2020, pp. 1166-1178.](https://doi.org/10.1515/jisys-2018-0299)

```python
# Vectorize the text using CountVectorizer
vectorizer = CountVectorizer(max_features=1000, stop_words='english')
X = vectorizer.fit_transform(Stephanie['cleaned_preprocessed_comment'])

# Apply Latent Dirichlet Allocation (LDA)
num_topics = 20  
lda = LatentDirichletAllocation(n_components=num_topics, random_state=0)
lda.fit(X)

# Display the top words in each topic
feature_names = vectorizer.get_feature_names_out()
for topic_idx, topic in enumerate(lda.components_):
    top_words = [feature_names[i] for i in topic.argsort()[:-10 - 1:-1]]
    print(f"Topic {topic_idx + 1}: {' '.join(top_words)}")
```

# Aspect Extraction 

```python
# Extracted topic keywords
list_of_topic_keywords = []

for topic_idx, topic in enumerate(lda.components_):
    top_keywords_idx = topic.argsort()[:-10 - 1:-1]  # Top 10 keywords per topic
    top_keywords = [feature_names[i] for i in top_keywords_idx]
    list_of_topic_keywords.append(top_keywords)

# Perform aspect extraction
aspects = []

for topic_idx, topic_keywords in enumerate(list_of_topic_keywords):
    aspect = {
        "Aspect": f"Aspect {topic_idx + 1}",
        "Keywords": topic_keywords,
        "Context": []
    }
    aspects.append(aspect)

for aspect in aspects:
    aspect_keywords = aspect["Keywords"]
    relevant_context = []

    for sentence in Stephanie['cleaned_preprocessed_comment']:
        if any(keyword in sentence for keyword in aspect_keywords):
            relevant_context.append(sentence)

    aspect["Context"] = relevant_context

# Display and analyze the extracted aspects
for aspect in aspects:
    print(f"Aspect: {aspect['Aspect']}")
    print(f"Keywords: {', '.join(aspect['Keywords'])}")
    print("Context:")
    for i, context in enumerate(aspect['Context']):
        if i < 5:  # Display only the first 5 sentences
            print(f"- {context}")
        else:
            print("... (More sentences not displayed)")
            break
    print("\n")
```

# Sentiment Analysis 

```python
for aspect in aspects:
    sentiment_scores = []
    for sentence in aspect['Context']:
        blob = TextBlob(sentence)
        sentiment_scores.append(blob.sentiment.polarity)
    aspect['SentimentScores'] = sentiment_scores
```

# Visualizations 

```python
for aspect in aspects:
    plt.figure(figsize=(8, 6))
    plt.bar(['Positive', 'Neutral', 'Negative'], 
            [sum(score > 0 for score in aspect['SentimentScores']),
             sum(score == 0 for score in aspect['SentimentScores']),
             sum(score < 0 for score in aspect['SentimentScores'])])
    plt.title(f"Sentiment Distribution for {aspect['Aspect']}")
    plt.xlabel('Sentiment')
    plt.ylabel('Count')
    plt.show()
```
